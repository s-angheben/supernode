```python
import pandas as pd
df = pd.read_csv("results.txt", sep="|", skipinitialspace=True)
# Rename the columns by stripping spaces
df.columns = [col.strip() for col in df.columns]

# Remove any leading/trailing spaces from the data in each cell
df = df.map(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
df
```

```python
df['Basic Graphs (60)'] = pd.to_numeric(df['Basic Graphs (60)'], errors='coerce')
# Divide by 60 only if the value is a number
df['Basic Graphs (60) Accuracy'] = df['Basic Graphs (60)'].apply(lambda x: x / 60 if pd.notnull(x) else x)

numeric_columns = ['Regular Graphs (100)', '4-vertex (20)', 'Distance_Regular (20)']
for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Perform arithmetic operations
df['Regular Graphs all (140)'] = df['Regular Graphs (100)'].fillna(0) + \
                                  df['4-vertex (20)'].fillna(0) + \
                                  df['Distance_Regular (20)'].fillna(0)

# Divide by 60 only if the value is a number
df['Regular Graphs all (140) Accuracy'] = df['Regular Graphs all (140)'].apply(lambda x: x / 140 if pd.notnull(x) else x)

df['CFI Graphs (100)'] = pd.to_numeric(df['CFI Graphs (100)'], errors='coerce')
# Divide by 60 only if the value is a number
df['CFI Graphs (100) Accuracy'] = df['CFI Graphs (100)'].apply(lambda x: x / 100 if pd.notnull(x) else x)

df['Extension Graphs (100)'] = pd.to_numeric(df['Extension Graphs (100)'], errors='coerce')
# Divide by 60 only if the value is a number
df['Extension Graphs (100) Accuracy'] = df['Extension Graphs (100)'].apply(lambda x: x / 100 if pd.notnull(x) else x)

df['all'] = df['Regular Graphs (100)'].fillna(0) + \
            df['4-vertex (20)'].fillna(0) + \
            df['Distance_Regular (20)'].fillna(0) + \
            df['CFI Graphs (100)'].fillna(0) + \
            df['Basic Graphs (60)'] + \
            df['Extension Graphs (100)']

df['all Accuracy'] = df['all'].apply(lambda x: x / 400 if pd.notnull(x) else x)
```

```python
column_order = ['models', 'Basic Graphs (60)', 'Basic Graphs (60) Accuracy', 'Regular Graphs all (140)', 'Regular Graphs all (140) Accuracy',
                'Extension Graphs (100)','Extension Graphs (100) Accuracy','CFI Graphs (100)','CFI Graphs (100) Accuracy',
                'all', 'all Accuracy']#, 'Total (400)', 'Accuracy']
df = df.reindex(columns=column_order)
```

```python
# Custom print each row
output = []
for index, row in df.iterrows():
    name = row['models']
    values = []
    for i in range(1, len(row), 2):  # Iterate over every second column starting from index 1
        value = row[i]  # Get value from the row
        accuracy = row[i + 1] * 100 if not pd.isnull(row[i + 1]) else float('nan')  # Get accuracy from the row and convert to percentage
        value_str = int(value) if not pd.isnull(value) else 'NaN'  # Handle NaN values for value
        acc_str = f"{accuracy:.1f}\%" if not pd.isnull(accuracy) else 'NaN'  # Handle NaN values for accuracy
        values.append(f"{value_str} & {acc_str}")  # Format value and accuracy
    output.append(f"{name} & {' & '.join(values)} \\\\")

# Print the output
for line in output:
    print(line)
    print("\hline")
```
