```python
import matplotlib.pyplot as plt
import torch
from torch_geometric.nn import MessagePassing
from torch_geometric.utils import add_self_loops, degree
import numpy as np
import networkx as nx
from torch_geometric.utils import from_networkx, to_networkx

def visualize(g, transf):
    if transf:
        ntype = set(nx.get_node_attributes(g, "ntype").values())
        ntype.remove('ORIG')
        
        nx.draw_networkx(G=g, pos=nx.spring_layout(g, np.random.seed(1234)), 
                     node_color=['#1f78b4' if n=='ORIG' else '#1fffff' for n in nx.get_node_attributes(g, "ntype").values()],
                     #labels =  nx.get_node_attributes(g, "ntype")   
                    )
    else:
        nx.draw_networkx(G=g, pos=nx.spring_layout(g, np.random.seed(1234)))

def showConcepts(g, k, v):
    N = list(g.neighbors(k))
    N.append(k)
    
    print(list(g.neighbors(k)), v, k)
    nx.draw_networkx(G=g, pos=nx.spring_layout(g, np.random.seed(1234)), 
                 node_color=['#1fffff' if n in N else '#1f78b4' for n in nx.nodes(g)]                    )
    plt.show()

```

```python
import torch
import torch_geometric
import itertools
import numpy as np


B = 1  # batch size
in_channels = 1
out_channels = 8
K = 2   # Number of hops
node_dim = 1    # data is in form [batch_size, num_nodes, num_features]
conv = torch_geometric.nn.TAGConv(in_channels=in_channels, out_channels=out_channels, K=K, node_dim=node_dim) # Graph convolution: 'ChebConv', 'TAGConv', 'SGConv',

# create a graph which consists of 2 hexagone which are linked with one edge between not 0 and 9
edge_index = list(itertools.permutations(range(0,6), 2)) + list(itertools.permutations(range(6,12), 2)) + [(0,9), (9,0)]
edge_index = np.transpose(edge_index)
edge_index = torch.from_numpy(edge_index)
# print("edge_index=", edge_index)
N= edge_index.max().item() + 1

interested_nodes = torch.tensor([1,2])
subset, sub_edge_index, mapping, _ = torch_geometric.utils.k_hop_subgraph(interested_nodes, edge_index=edge_index, num_hops=K, relabel_nodes=True)

x= torch.randn(B, N, in_channels)

# print("x=", x)
# print("subset=", subset)
# print("sub_edge_index=", sub_edge_index)

out1 = conv(x, edge_index=edge_index)
# print("out1 (before)=", out1)
out1 = out1.index_select(node_dim, interested_nodes)

out2 = conv(x.index_select(node_dim, subset), sub_edge_index)
# print("out2 (before)=", out2)
out2 = out2.index_select(node_dim, mapping)
print("diff (out2-out1)=", out2-out1)
print("out1=", out1)
print("out2=", out2)
```
